```python
def sigmoid(x):
    return 1/(1+np.exp(-x))



def gradient_descent(age, sex, y, epoch):
    w1_curr = w2_curr = 1
    bias = 0
    lr = 0.05
    n = len(y)
    
    
    
    #
    
    
    for i in range(epoch):
        weighted_sum = w1_curr*age + w2_curr*sex + bias
        yp = sigmoid(weighted_sum)
        cost = (1/n) * sum([val**2 for val in (y-yp)])
        
        
        
        #now taking the derivatives
        w1d = -(2/n)*sum(w1_curr*(y-yp))
        w2d = -(2/n)*sum(w2_curr*(y-yp))
        bd = -(2/n)*sum(y-yp)
        #bd = (1/n)*sum([k for k in (yp-y)])
        
        
        #alternating the coefficients and the bias term
        
        w1_curr = w1_curr - lr*w1d
        w2_curr = w2_curr - lr*w2d
        bias = bias - lr*bd
        

        
        print(f'epoch:{i}, w1:{w1_curr}, w2:{w2_curr}, bias:{bias}, loss:{cost}')
```
